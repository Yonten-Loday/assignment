```python
import random
```

```python
aluminum = np.array([6.4e-5, 3.01e-5, 2.36e-5, 3.0e-5, 7.0e-5, 4.5e-5, 3.8e-5, 4.2e-5, 2.62e-5,3.6e-5])
copper = np.array([4.5e-5, 1.97e-5, 1.6e-5, 1.97e-5, 4.0e-5, 2.4e-5, 1.9e-5, 2.41e-5, 1.85e-5,3.3e-5])
steel = np.array([3.3e-5, 1.2e-5, 0.9e-5, 1.2e-5, 1.3e-5, 1.6e-5, 1.4e-5, 1.58e-5, 1.32e-5,2.1e-5])

material_x = []

for i in range(0,10):
    material_x.append(random.random()*10**-4)
```

```python
material_x
```

```python
# Calculate the average
aluminum_mean = np.mean(aluminum)
copper_mean = np.mean(copper)
steel_mean = np.mean(steel)

material_x_mean =np.mean(material_x)
```

```python
# Create arrays for the plot
materials = ['Aluminum', 'Copper', 'Steel', 'Material_x']
x_pos = np.arange(len(materials))
CTEs = [aluminum_mean, copper_mean, steel_mean, material_x_mean]
```

```python
# Build the plot
fig, ax = plt.subplots()

ax.bar(x_pos, CTEs, align='center', alpha=0.5)
ax.set_ylabel('Coefficient of Thermal Expansion (\degree C1)')
ax.set_xticks(x_pos)
ax.set_xticklabels(materials)
ax.set_title('Coefficient of Thermal Expansion (CTE)')
ax.yaxis.grid(True)
#ax.xaxis.grid(True)

# Save the figure and show
plt.tight_layout()
plt.savefig('bar_plot.png')
plt.show()
```

```python
df = pd.read_csv('fishdata.csv', encoding='unicode_escape', header=None)
```

```python
mylist = [*range(12,35)]
fish = df.drop(columns=mylist)
fish
```

```python
header = ['year', 'month', 'day', 'hour', 'minute', 'second', 'carbon monoxide', 'PM 2.5', 'external temp', 'humidity', 'internal temperature', 'ignore']
fish.columns = header
fish
```

```python
fish = fish.dropna()
fish
```

```python
fish.dtypes
```

```python
fish['year'] = fish['year'].astype('int64')
fish['month'] = fish['month'].astype('int64')
fish['minute'] = fish['minute'].astype('int64')
fish['second'] = fish['second'].astype('int64')
fish.dtypes
```

```python
fish['carbon monoxide'] = pd.to_numeric(fish['carbon monoxide'], errors='coerce')
fish['carbon monoxide'] = fish['carbon monoxide'].astype('float64')
fish.dtypes
```

```python
fish.describe()
```

```python
fish = fish[fish['year'] > 2000]
fish = fish[fish['year'] <= 2021]
fish.describe()
```

```python
fish = fish[fish['external temp'] > 0]
fish.describe()
```

```python
fish_modi = fish.drop(['year', 'PM 2.5', 'humidity', 'ignore'], axis='columns')
fish_modi.describe()
```

```python
fish_modi['time_code'] = range(len(fish_modi))
```

```python
x = fish_modi['time_code']
y1 = fish_modi['external temp']
y2 = fish_modi['internal temperature']

fig, ax = plt.subplots(2, figsize=(12, 12), dpi=80)
ax[0].plot(x, y1, 'r')
ax[1].plot(x, y2, 'g')
ax[0].set_title('temperature record', fontsize=30)
ax[0].set_xlabel('time_code', fontsize=10)
ax[0].set_ylabel('external temp', fontsize=10)
ax[1].set_xlabel('time_code', fontsize=10)
ax[1].set_ylabel('internal temp', fontsize=10)
ax[0].yaxis.grid(True)
ax[1].yaxis.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,6)) 
plt.plot(x, y1, color='r', label='external')
plt.plot(x, y2, color='g', label='internal')
plt.xlabel('time_code')
plt.ylabel('temperature')
plt.legend()
plt.grid(axis='y')
plt.show()
```

```python
labels = ['Very High', 'High', 'Middle', 'Low', 'Very Low']
fish_modi['ex_rank'] = pd.qcut(fish_modi['external temp'], q=5 ,labels=labels)
fish_modi['in_rank'] = pd.qcut(fish_modi['internal temperature'], q=5 ,labels=labels)
fish_modi.head()
```

```python
plt.hist(fish_modi['external temp'], bins=5, color='r')
```

```python
plt.hist(fish_modi['internal temperature'], bins=5, color='g')
```

```python
legend = ['external temperature', 'internal temperature']
x_pos = np.arange(len(legend))
average = [fish_modi['external temp'].mean(), fish_modi['internal temperature'].mean()]
```

```python
fig, ax = plt.subplots()
ax.bar(x_pos, average, align='center', color=['r', 'g'], alpha=0.5)
ax.set_ylabel('Temperature')
ax.set_xticks(x_pos)
ax.set_xticklabels(legend)
ax.set_title('Temperature record')
ax.yaxis.grid(True)
plt.show()
```

```python
plt.bar(labels, fish_modi.groupby('ex_rank').count()['external temp'], color='r')
plt.title('External temperature')
plt.grid(axis='y')
plt.show()
```

```python
plt.bar(labels, fish_modi.groupby('in_rank').count()['internal temperature'], color='g')
plt.title('Internal temperature')
plt.grid(axis='y')
plt.show()
```

```python
fish_pivot = pd.pivot_table(fish_modi, values=['external temp', 'internal temperature'], 
                            index=['month'], aggfunc=np.mean)
fish_pivot
```

```python
x = fish_pivot.index
y1 = fish_pivot['external temp']
y2 = fish_pivot['internal temperature']

fig, ax = plt.subplots(2, figsize=(8, 8), dpi=80)
ax[0].plot(x, y1, color='r', marker='o')
ax[1].plot(x, y2, color='g', marker='o')
ax[0].set_title('temperature record', fontsize=30)
ax[0].set_xlabel('time_code', fontsize=10)
ax[0].set_ylabel('external temp', fontsize=10)
ax[1].set_xlabel('time_code', fontsize=10)
ax[1].set_ylabel('internal temp', fontsize=10)
ax[0].yaxis.grid(True)
ax[1].yaxis.grid(True)
plt.show()
```
